```python
#!g1.1
import pandas as pd
df = pd.read_csv('data/Certificate_2020_ds_0.txt', delimiter=";")
```

```python
#!g1.1
markup = pd.read_csv('markup.csv', delimiter=";")
```

```python
#!g1.1
import re
array = []
for id in markup['Номер эВСД оформленного с нарушениями']:
    id = str(id).replace('\n', ',').replace(';', ',')
    id = re.sub(r'[^\d, ]+', '', id).replace(' ', ',')
    id = id.split(',')
    #print(id)
    for i in id:
        if i != '' and len(i) > 5:
            # print(i)
            array.append(i)
```

```python
#!g1.1
import time

t = round(time.time() * 1000)
sus = []
for idx, id in enumerate(df['id']):
    id = str(id)
    if id in array:
        print(idx, id, 'sus', round(time.time() * 1000) - t)
        sus.append(True)
        continue
    sus.append(False)
df['sus'] = sus
```

```python
#!g1.1
df.to_csv('Certificate_2020_ds_0.csv')
```

```python
#!g1.1
reasons = []
for index, row in df.iterrows():
    if index != len(reasons):
        print(row['id'], len(reasons), index)
    if row['sus'] == True:
        for idx, m in enumerate(markup['Номер эВСД оформленного с нарушениями']):
            m = str(m).replace('\n', ',').replace(';', ',')
            m = re.sub(r'[^\d, ]+', '', m).replace(' ', ',')
            m = m.split(',')
            if str(row['id']) in m:
                reasons.append(markup['Вид нарушения'][idx])
                break
    else:
        reasons.append('')
    if index != len(reasons) - 1:
        print(row['id'], len(reasons), index)
    

print(len(df['id']), len(reasons))
df['reasons'] = reasons
```

```python
#!g1.1
df.to_csv('cert_0.csv')
```
